Prepare raw data for analysis and modeling. Do not perform scaling or encoding here.

1ï¸âƒ£ Imports & Setup

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
```

2ï¸âƒ£ Load Raw Data

```python
df = pd.read_csv('../data/raw/data.csv')
df.head()
```

3ï¸âƒ£ Initial Inspection

```python
df.shape
df.info()
df.describe(include='all')
```

4ï¸âƒ£ Handle Missing Values

```python
# Numeric â†’ median
num_cols = df.select_dtypes(include=np.number).columns
for col in num_cols:
    df[col].fillna(df[col].median(), inplace=True)

# Categorical â†’ mode
cat_cols = df.select_dtypes(include='object').columns
for col in cat_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

5ï¸âƒ£ Remove Duplicates

```python
df.drop_duplicates(inplace=True)
```

6ï¸âƒ£ Fix Data Types

```python
# Example
# df['date'] = pd.to_datetime(df['date'])
```

7ï¸âƒ£ Outlier Detection (IQR Method)

```python
for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    df = df[(df[col] >= Q1 - 1.5*IQR) & (df[col] <= Q3 + 1.5*IQR)]
```

8ï¸âƒ£ Save Cleaned Data

```python
df.to_csv('../data/interim/cleaned.csv', index=False)
```
