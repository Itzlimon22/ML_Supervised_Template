Create model-meaningful features.

1Ô∏è‚É£ Load Cleaned Data

```python
import pandas as pd

df = pd.read_csv('../data/interim/cleaned.csv')
```

2Ô∏è‚É£ Separate Target

```python
target = 'target_column'
X = df.drop(target, axis=1)
y = df[target]
```

3Ô∏è‚É£ Encoding Categorical Features

```python
from sklearn.preprocessing import OneHotEncoder

X = pd.get_dummies(X, drop_first=True)
```

4Ô∏è‚É£ Feature Selection (Correlation Based)

```python
corr = pd.concat([X, y], axis=1).corr()[target].abs()
important_features = corr[corr > 0.1].index
X = X[important_features.drop(target)]
```

5Ô∏è‚É£ Save Engineered Features

```python
processed = pd.concat([X, y], axis=1)
processed.to_csv('../data/processed/features.csv', index=False)
```
